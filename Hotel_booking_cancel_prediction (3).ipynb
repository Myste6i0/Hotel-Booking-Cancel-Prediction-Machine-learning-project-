```python
#  chi-squared test of independence for each categorical variable in a DataFrame (df) concerning its relationship with the 'is_canceled' variable
from scipy.stats import chi2_contingency # Import the chi-squared test function from the scipy.stats module.
cols= df.columns
for col in cols:
  contingency_table = pd.crosstab(df['is_canceled'], df[col]) # cross-tabulation or crosstab
  chi2, p, _, _ = chi2_contingency(contingency_table) # Perform chi-squared test
  # the chi2 variable holds the chi-squared statistic, and the p variable holds the p-value
  if p < 0.05:
      print(f"There is a significant relationship between 'is_canceled' and '{col}'.")
  else:
      print(f"There is no significant relationship between 'is_canceled' and '{col}'.")
```

```python
drop_col = []
is_canceled_correlations = df.corr()['is_canceled']
for col, correlation in is_canceled_correlations.items():
    if abs(correlation) < 0.1 and col != 'is_canceled':
        drop_col.append(col)
```

```python
for col in columns:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  IQR = q3 - q1
  lower_limit = q1 - 1.5 *IQR
  upper_limit = q3 + 1.5 *IQR
  df[col] = df[col].clip(lower=lower_limit,upper=upper_limit)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=44,test_size=0.20)
```

```python
logestic_regression = LogisticRegression()
logestic_regression.fit(x_train,y_train)
lr_y_pred = logestic_regression.predict(x_test)
print(accuracy_score(y_test,lr_y_pred))
```

```python
kn = KNeighborsClassifier()
kn.fit(x_train,y_train)
lr_y_pred = kn.predict(x_test)
print(accuracy_score(y_test,lr_y_pred))
```

```python
svc = SVC()
svc.fit(x_train,y_train)
lr_y_pred = svc.predict(x_test)
print(accuracy_score(y_test,lr_y_pred))
```

```python
dfc = DecisionTreeClassifier()
dfc.fit(x_train,y_train)
lr_y_pred = dfc.predict(x_test)
print(accuracy_score(y_test,lr_y_pred))
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
lr_y_pred = rf.predict(x_test)
print(accuracy_score(y_test,lr_y_pred))
```
